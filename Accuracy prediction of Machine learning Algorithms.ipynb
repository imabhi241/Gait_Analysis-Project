```python
import pandas as pd
import numpy as np
```

```python
datasix1 = pd.read_csv(r'C:\Users\hp\Desktop\Major project (Final year)_24\DATA\Human Gait Phase Dataset [Master Data]\data\GP1_0.6_marker.csv')
dataseven2 = pd.read_csv(r'C:\Users\hp\Desktop\Major project (Final year)_24\DATA\Human Gait Phase Dataset [Master Data]\data\GP1_0.7_marker.csv')
datasix1.head()
```

```python
data1 = pd.DataFrame({'secim':np.zeros(12000)})
data2 = pd.DataFrame({'secim':np.ones(12000)})

data2 = pd.concat([dataseven2,data2],axis=1)
data1 = pd.concat([datasix1,data1],axis=1)

data = data1.append(data2,ignore_index=True)
data.head()
```

```python
x = data.drop(["secim"],axis=1)
y = data.secim.values
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC

svm = SVC(random_state = 1)
svm.fit(x_train,y_train)
print("Accuracy of svm is :",int(svm.score(x_test,y_test)*100),"%")
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train,y_train)
print('Accuracy of Bayes in test data is :', int(nb.score(x_test,y_test)*100),"%")
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print('Accuracy of decision tree in test data is:',int(dt.score(x_test,y_test)*100),"%")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100,random_state=1)
rf.fit(x_train,y_train)
print('Accuracy of Random Forest on test data is : ',int(rf.score(x_test,y_test)*100),"%")
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
print("test accuracy for Log Regressin is  {}".format(lr.score(x_test,y_test)*100),"%")
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3) #n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)

print("k={} nn Accuracy:{}".format(3,int(knn.score(x_test,y_test))*100),"%")
```
